```python
import pandas as pd
import plotly.express as px
import statistics

df = pd.read_csv("C:/Users/ADMIN/Desktop/WhiteHat_Jr/General/Data/savings_data_final.csv")
```

```python
fig = px.scatter(df,y="quant_saved",color="rem_any")
fig.show()
```

```python
q1 = df["quant_saved"].quantile(0.25)
q3 = df["quant_saved"].quantile(0.75)
iqr = q3 - q1
lower_limit = q1 - 1.5*iqr
upper_limit = q3 + 1.5*iqr

print("Quantile 1:",q1)
print("Quantile 3:",q3)
print("Inter Quantile Range:",iqr)
print("Lower Limit:",lower_limit)
print("Upper Limit:",upper_limit)
```

```python
new_df = df[df["quant_saved"] < upper_limit]
all_savings = new_df["quant_saved"].tolist()
```

```python
mean = statistics.mean(all_savings)
median = statistics.median(all_savings)
mode = statistics.mode(all_savings)
std = statistics.stdev(all_savings)

print("Mean:",mean)
print("Median:",median)
print("Mode:",mode)
print("Standard Deviation:",std)
```

```python
import plotly.figure_factory as ff

fig = ff.create_distplot([all_savings],["Savings"],show_hist=False)
fig.show()
```

```python
import random
import plotly.graph_objects as go

sampling_data = []

for x in range(1000):
    temp_data = []
    for i in range(100):       
        random_index = random.randint(0,len(all_savings)-1)
        temp_value = all_savings[random_index]
        temp_data.append(temp_value)
    temp_mean = statistics.mean(temp_data)
    sampling_data.append(temp_mean)
    
final_mean = statistics.mean(sampling_data)

fig = ff.create_distplot([sampling_data],["Sampling Savings"],show_hist=False)
fig.add_trace(go.Scatter(x=[final_mean,final_mean], y = [0,0.08], mode="lines", name="Mean"))
fig.show()
```

```python
import numpy as np

# Population Correlation
population_temp_df = new_df[new_df.age != 0]

population_age = population_temp_df["age"].tolist()
population_savings = population_temp_df["quant_saved"].tolist()
```

```python
print("population_std",statistics.stdev(all_savings))
print("population_mean",statistics.mean(all_savings))
print("population_median",statistics.median(all_savings))
print("population_mode",statistics.mode(all_savings))
print("population_correlation",np.corrcoef(population_age, population_savings)[0,1])

print("------------------------------------------------------------------------------")

print("sampling_std",statistics.stdev(sampling_data))
print("sampling_mean",statistics.mean(sampling_data))
print("sampling_median",statistics.median(sampling_data))
print("sampling_mode",statistics.mode(sampling_data))
```

```python
reminded_df = new_df.loc[new_df["rem_any"] == 1]
not_reminded_df = new_df.loc[new_df["rem_any"] == 0]




not_reminded_savings = not_reminded_df["quant_saved"].tolist()

sampling_mean_list_not_reminded = []
for i in range(1000):
  temp_list = []
  for j in range(100):
    temp_list.append(random.choice(not_reminded_savings))
  sampling_mean_list_not_reminded.append(statistics.mean(temp_list))

mean_sampling_not_reminded = statistics.mean(sampling_mean_list_not_reminded)
stdev_sampling_not_reminded = statistics.stdev(sampling_mean_list_not_reminded)



reminded_savings = reminded_df["quant_saved"].tolist()

sampling_mean_list_reminded = []
for i in range(1000):
  temp_list = []
  for j in range(100):
    temp_list.append(random.choice(reminded_savings))
  sampling_mean_list_reminded.append(statistics.mean(temp_list))

mean_sampling_reminded = statistics.mean(sampling_mean_list_reminded)
stdev_sampling_reminded = statistics.stdev(sampling_mean_list_reminded)


z_score = (mean_sampling_reminded - mean_sampling_not_reminded) / stdev_sampling_not_reminded
```
